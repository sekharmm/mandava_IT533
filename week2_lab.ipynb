```python
bins = np.linspace(amazon.number_of_reviews.min(),amazon.number_of_reviews.max(),4)
bins
```

```python
labels = ['small', 'medium', 'high']
```

```python
amazon3=amazon
amazon3['bins'] = pd.cut(amazon3['number_of_reviews'], bins=bins, labels=labels, include_lowest=True)
```
