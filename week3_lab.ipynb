```python
amazon = pd.read_csv("https://raw.githubusercontent.com/sekharmm/mandava_IT533/main/amazon_toy_sample_clean.csv", encoding="ISO-8859-1") # use your own GitHub link here
amazon.dropna(inplace=True)
bins = np.linspace(amazon.number_of_reviews.min(),amazon.number_of_reviews.max(),4)
bins
```

```python
labels = ['small', 'medium', 'high']
```

```python
amazon3=amazon
amazon3['bins'] = pd.cut(amazon3['number_of_reviews'], bins=bins, labels=labels, include_lowest=True)
```

```python
amazon3.head()
```

```python
amazon3[amazon3.number_of_reviews>= 1000]
```

```python
amazon3.groupby(['bins']).count()
```

```python
plt.hist(amazon3['bins'], bins=3)
```
